# Mesh and volume

Voxel-based volumes and triangle-based meshes can exist together in the same scene.

```python
import ipyniivue
from ipyniivue import NiiVue, SliceType, download_dataset

nv = NiiVue(slice_type=SliceType.MULTIPLANAR)
nv.load_volumes([{"path": "./images/mni152.nii.gz"}])
nv.load_meshes([{
            "path": "./images/BrainMesh_ICBM152.lh.mz3",
            "rgba255": [128, 192, 255, 255],
        }])


nv
```
