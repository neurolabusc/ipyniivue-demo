# Time Series

This Python code is similar to the [time series web page](https://niivue.com/demos/features/timeseries2.html)

```python
import ipywidgets as widgets
from ipyniivue import NiiVue, SliceType, download_dataset
from ipyniivue.download_dataset import DATA_FOLDER


nv = NiiVue(slice_type=SliceType.MULTIPLANAR)
nv.load_volumes([{"path": "./images/mpld_asl.nii.gz"}])


# nv.opts.multiplanarShowRender = niivue.SHOW_RENDER.ALWAYS
nv.graph.auto_size_multiplanar = True
nv.graph.normalize_values = False
nv.graph.opacity = 1.0
nvols = 4 #TODO nv.volumes[0].n_frame_4d
slider = widgets.IntSlider(min=0, max=nvols - 1, description="Volume")


def update_frame(*args):
    """Select the frame corresponding to the slider value."""
    nv.volumes[0].frame_4d = slider.value


slider.observe(update_frame, "value")

display(slider)
display(nv)
```
