# ipyniivue — Minimal demo

This notebook demonstrates the minimal usage of `NiiVue` with a single local NIfTI file.


```python
import ipyniivue
nv = NiiVue(
    show_3d_crosshair=True, back_color=(1, 1, 1, 1), mesh_xray=0.3, is_colorbar=True
)
mesh_layer = {
    "path": "./images/BrainMesh_ICBM152.lh.motor.mz3",
    "cal_min": 0.5,
    "cal_max": 5.5,
    "use_negative_cmap": True,
    "opacity": 0.7,
}
nv.load_meshes(
    [
        {
            "path": "./images/BrainMesh_ICBM152.lh.mz3",
            "rgba255": [255, 255, 255, 255],
            "layers": [mesh_layer],
        },
        {"path":  "./images/CIT168.mz3", "rgba255": [0, 0, 255, 255]},
    ]
)
nv.opts.show_legend = False
# nv.set_mesh_layer_property(nv.meshes[0].id, 0, "colorbar_visible", False)
nv
```
